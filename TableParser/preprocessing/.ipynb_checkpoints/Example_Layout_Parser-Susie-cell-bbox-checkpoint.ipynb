````python
def get_cell_coord_DocParser(LayoutParser_TextBlock):
    '''param LayoutParser_TextBlock is one element in the layout generated by Google Vision API.
    E.g., 
    TextBlock(block=Quadrilateral(points=[[399   8]
     [425   7]
     [426  24]
     [400  25]], width=27, height=18), text=243, id=21, type=None, parent=None, next=None, score=0.9599999785423279)
     
     Note that the sequence of points is upper left, upper right, lower right, lower left.
     '''
    points = LayoutParser_TextBlock.points
    
    x_upper_left = points[0][0]
    y_upper_left = points[0][1]
    
    x_upper_right = points[1][0]
    y_upper_right = points[1][1]
    
    x_lower_right = points[2][0]
    y_lower_right = points[2][1]
    
    x_lower_left = points[3][0]
    y_lower_left = points[3][1]
    
    x_max_left = max(x_upper_left, x_lower_left)
    x_max_right = max(x_upper_right, x_lower_right)
    y_max_upper = max(y_upper_left, y_upper_right)
    y_max_lower = max(y_lower_left, y_lower_right)
    
    text = LayoutParser_TextBlock.text
    width = LayoutParser_TextBlock.width
    height = LayoutParser_TextBlock.height
    
    return (text, [x_max_left, y_max_upper, width, height])
````

### getting the name column

````python
# extract name of row 

filtered_name = layout.filter_by(
        lp.Rectangle(x_1=1, y_1=0, x_2=375, y_2=1175),soft_margin = {"left":5, "right":5}
    )

blocks = filtered_name

blocks = sorted(blocks, key = lambda x: x.coordinates[1])
    # Sort the blocks vertically from top to bottom
distances = np.array([b2.coordinates[1] - b1.coordinates[3] for (b1, b2) in zip(blocks, blocks[1:])])
    # Calculate the distances:
    # y coord for the upper edge of the bottom block -
    #   y coord for the bottom edge of the upper block
    # And convert to np array for easier post processing
plt.hist(distances, bins=50);
plt.axvline(x=3, color='r');
    # Let's have some visualization
distance_th = 0

distances = np.append([0], distances) # Append a placeholder for the first word
block_group = (distances>distance_th).cumsum() # Create a block_group based on the distance threshold

block_group

# Group the blocks by the block_group mask
grouped_blocks = [[] for i in range(max(block_group)+1)]
for i, block in zip(block_group, blocks):
    grouped_blocks[i].append(block)

def group_blocks_by_distance(blocks, distance_th):

    blocks = sorted(blocks, key = lambda x: x.coordinates[1])
    distances = np.array([b2.coordinates[1] - b1.coordinates[3] for (b1, b2) in zip(blocks, blocks[1:])])

    distances = np.append([0], distances)
    block_group = (distances>distance_th).cumsum()

    grouped_blocks = [lp.Layout([]) for i in range(max(block_group)+1)]
    for i, block in zip(block_group, blocks):
        grouped_blocks[i].append(block)

    return grouped_blocks
````

````python
A = group_blocks_by_distance(filtered_name, -10)
height_th = 1175
# And finally we combine the outputs

idxA = 1
result = []

while idxA < len(A) :
    ay = A[idxA][0].coordinates[1]
    ares = ''.join(A[idxA].get_texts())
    if ay  < height_th:
        idxA += 1
    else:
        ares = ''
    result.append([ares])

name = result
name = pd.DataFrame (name,columns=['name'])

name
````

#### expanding the cells

````python
temp_dict = defaultdict(lambda: defaultdict(str))
for i, content in enumerate(filtered_name):
    out = get_cell_coord_DocParser(content)
    '''print('id:', id_start)
    print('bbox:', out[1])
    print('text:', out[0])
    print('properties:', '{}-{},{}-{}'.format(i,i,col_counter,col_counter))
    print('row_range:', [i,i])
    print('col_range:', [col_counter,col_counter])
    print('\n')'''
    
    temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['id'] = id_start
    temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['bbox'] = out[1]
    temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['text'] = out[0]
    temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['row_range'] = [i,i]
    temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['col_range'] = [col_counter,col_counter]
    
    id_start += 2
    
````

````python
max_cell_width = max([d['bbox'][2] for d in temp_dict.values()])
min_x_left = min([d['bbox'][0] for d in temp_dict.values()])

for _, d in temp_dict.items():
    row_range = d['row_range']
    bbox = d['bbox']
    cell_bboxes_for_row = bbox
    bbox[2] = max_cell_width
    bbox[0] = min_x_left
    temp_dict[_]['bbox'] = bbox
previous_x_left = min_x_left
````

````python
col_counter = 0

len_filtered_tem = len(filtered_name)
for properties, d in temp_dict.items(): 
    skeleton.append( {
                        "id": d["id"],
                        "category": "table_cell",
                        "properties": properties,
                        "row_range": d["row_range"],
                        "col_range": d["col_range"],
                        "parent": 7
                      })
    skeleton.append( {
                        "id": d["id"]+1,
                        "category": "box",
                        "page": 0,
                        "bbox": d["bbox"],
                        "text": d["text"],
                        "parent": d["id"]
                      })

col_counter += 1
````

### getting each data column

````python
def out_skeleton():
    global filtered_tem, col_counter, id_start, previous_x_left
    temp_dict = defaultdict(lambda: defaultdict(str))
    for i, content in enumerate(filtered_tem):
        out = get_cell_coord_DocParser(content)
        print('id:', id_start)
        print('bbox:', out[1])
        print('text:', out[0])
        print('properties:', '{}-{},{}-{}'.format(i,i,col_counter,col_counter))
        print('row_range:', [i,i])
        print('col_range:', [col_counter,col_counter])
        print('\n')

        temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['id'] = id_start
        temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['bbox'] = out[1]
        temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['text'] = out[0]
        temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['row_range'] = [i,i]
        temp_dict['{}-{},{}-{}'.format(i,i,col_counter,col_counter)]['col_range'] = [col_counter,col_counter]

        id_start += 2
    
    max_cell_width = max([d['bbox'][2] for d in temp_dict.values()])
    min_x_left = min([d['bbox'][0] for d in temp_dict.values()])

    for _, d in temp_dict.items():
        row_range = d['row_range']
        bbox = d['bbox']
        cell_bboxes_for_row = bbox
        bbox[2] = max_cell_width
        if previous_x_left + max_cell_width <= min_x_left:
            
            bbox[0] = min_x_left
        else:
            #bbox[0] = previous_x_left + max_cell_width
            bbox[0] = min_x_left
            
        temp_dict[_]['bbox'] = bbox
    
    previous_x_left = min_x_left
        

    len_filtered_tem = len(filtered_tem)
    for properties, d in temp_dict.items(): 
        skeleton.append( {
                            "id": d["id"],
                            "category": "table_cell",
                            "properties": properties,
                            "row_range": d["row_range"],
                            "col_range": d["col_range"],
                            "parent": 7
                          })
        skeleton.append( {
                            "id": d["id"]+1,
                            "category": "box",
                            "page": 0,
                            "bbox": d["bbox"],
                            "text": d["text"],
                            "parent": d["id"]
                          })
                        
    col_counter += 1
````

````python
import json
import numpy as np

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)
````
